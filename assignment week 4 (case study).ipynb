# 3.Bar graph of male vs female buyers participated in the sales

```python
ax=cars['Buyer Gender'].hist(figsize=(7,7))
plt.title("Male vs Female buyers in the sales")
for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

Inference:
    1. The number of female in sales is more than the number of males.
