# 3.Check whether all the zones generate the same amount of sales

#### Using One-way ANOVA :
###### Level of significance, alpha = 0.05
###### Null Hypothesis, H0 : Same amount of sale for the four Zones.
###### Alternate Hypothesis, Ha : Different amount of sale for the four zones.

```python
#impoting dataset
from scipy.stats import f_oneway
```

```python
#oneway anova calculation
f_oneway(d1['Zone - A'],d1['Zone - B'],d1['Zone - C'],d1['Zone - D'])
```

##### Conclusion :
Here, p-value(0.0011827601694503335) is less than the alpha value (0.05). Hence we reject the null hypotheis and it can be concluded that the four zones have different sale.
