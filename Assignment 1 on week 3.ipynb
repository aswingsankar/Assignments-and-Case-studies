```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading a data set to environment

```python
iris = pd.read_excel('iris.xls')
```

```python
iris
```

# Displaying the columns

```python
iris.columns
```

# Calculating mean of each column

```python
np.mean(iris)
```

# To check for null values

```python
iris.isnull().sum()
```

# Different type of Visualizations

# SL vs PL

```python
plt.figure(figsize=(8,8))
iris[['SL','PL']].plot.kde()
plt.title("SL VS PL",fontsize=16)
plt.xlabel('SL')
plt.ylabel('SW')
```

# SW vs PW

```python
plt.figure(figsize=(8,8))
iris[['SW','PW']].plot.kde()
plt.title("SW VS PW",fontsize=16)
plt.xlabel('SW')
plt.ylabel('PW')
```

# SW and SL vs Classification

```python
plt.figure(figsize=(8,10))
plt.subplot(2,1,1)
plt.scatter(iris['Classification'],iris['SW'],s=15,marker='^')
plt.title("SW VS Classification",fontsize=16)
plt.xlabel('Classification')
plt.ylabel('SW')

plt.subplot(2,1,2)
plt.scatter(iris['Classification'],iris['SL'],s=15,marker='^')
plt.title("SL VS Classification",fontsize=16)
plt.xlabel('Classification')
plt.ylabel('SL')
plt.tight_layout()
```

```python
plt.figure(figsize=(8,10))

plt.subplot(2,1,1)
plt.scatter(iris['Classification'],iris['PW'],s=15,marker='^')
plt.title("PW VS Classification",fontsize=16)
plt.xlabel('Classification')
plt.ylabel('PW')

plt.subplot(2,1,2)
plt.scatter(iris['Classification'],iris['PL'],s=15,marker='^')
plt.title("PL VS Classification",fontsize=16)
plt.xlabel('Classification')
plt.ylabel('PL')
plt.tight_layout()
```
